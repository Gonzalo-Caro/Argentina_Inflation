```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
archivo = 'csv_creados/Data_Argentina_Relative_Terms.csv'

df = pd.read_csv(archivo)
df.head(2)
```

```python
df = df[['year', 'GDP_Constant_gr', 'Velocity_Money_gr']]
```

```python
df.info()
```

```python
df['year'] = pd.to_datetime(df['year'])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.histplot(df['GDP_Constant_gr'], kde=True, color='blue')
plt.title('GDP Constant Growth 1961 to 2023')
plt.xlabel('GDP Constant Growth')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(df['GDP_Constant_gr'])
plt.title('Boxplot of GDP Constant')
plt.xlabel('GDP Consntant Growth')
plt.ylabel('Frequency')
plt.ylim(-0.2, 0.2)
plt.show()
```

```python
sns.histplot(df['Velocity_Money_gr'], kde=True, color='blue')
plt.title('Velocity of Money Growth 1961 to 2023')
plt.xlabel('Velocity of Money Growth')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(df['Velocity_Money_gr'])
plt.title('Boxplot of Velocity of Money')
plt.xlabel('Velocity of Money Growth')
plt.ylabel('Frequency')
plt.ylim(-1, 2.5)
plt.show()
```
